## Crime Incident Reports & Venues Data Analysis in Boston, Massachusetts

```python
import pandas as pd
import numpy as np
import geocoder
import wget
import matplotlib.pyplot as plt
import requests # library to handle requests

from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library
```

```python
# Read in csv files into dataframe
crime_file = 'boston_crimerate.csv'
bos_district = 'boston_district.csv'

df_district = pd.read_csv(bos_district)
df_district = df_district.drop(['Lat', 'Long'], axis=1)

df_crime = pd.read_csv(crime_file)
print(df_crime.shape)
df_crime.head()
```

```python
# Crime Incident Rate Data preprocessing
df_crime.drop(df_crime.columns[[0, 1, 3, 5, 6, 8, 9, 10, 11, 12, 13, 16]], axis=1, inplace=True)
print(df_crime.shape)
df_crime
```

```python
df_crime.dropna(inplace=True)
df_crime.columns = ['CATEGORY', 'DISTRICT', 'DATE', 'LAT', 'LONG']
df_crime['DATE'] = pd.to_datetime(df_crime['DATE'])
```

```python
df_crime.sort_values(by='DATE', ascending=False, inplace=True)
df_crime.reset_index(drop=True, inplace=True)

print(df_crime.shape)
df_crime
```

```python
# Lets only look at crime data in year 2019
mask = (df_crime['DATE'] >= '2019-1-1')
df_crime = df_crime.loc[mask]
print(df_crime.shape)
df_crime
```

```python
# Now lets merge district name with crime data frame
df = pd.merge(df_crime, df_district, on='DISTRICT')

# This is the pre-processed data frame. It contains all the crime data
# in association with the district name and category of the crime
df
```

### Analysis

In this step, we will output some graphs to take a look at which category has the highest number of crime incident reports, which district has the highest number of reported crime incidents, and the top 5 types of crime incidents in each district.
PART2:Then, we will move on to see the number of different types of venues in each district. We will then use k-means clustering to cluster the venues into different category. 

```python
# Top 10 highest number of crime categories
print('There are {} uniques categories.'.format(len(df['CATEGORY'].unique())))
crime_category = df.groupby('CATEGORY').count()
crime_category.sort_values(by='DISTRICT', ascending=False, inplace=True)
crime_top10 = crime_category['DISTRICT'].head(10)
# print(crime_top10)


crime_top10.plot(kind='barh', figsize=(10, 10), color='RED')
plt.xlabel('Number of Reported Crime Incidents')
plt.ylabel('Crime Indicent Categories')
plt.title('Top 10 Crime Incidents in Boston between 1/1/2019 and 9/')

for index, value in enumerate(crime_top10): 
    label = format(int(value), ',') # format int with commas
    
    # place text at the end of bar (subtracting 47000 from x, and 0.1 from y to make it fit within the bar)
    plt.annotate(label, xy=(value - 500, index - 0.10), color='white')

    
plt.show()
```

```python
# Crime Incident reports in each district: choropleth map for visualization
district_crime = df.groupby('DISTRICT').count()
district_crime.sort_values(by='CATEGORY', ascending=False, inplace=True)
district_crime.reset_index(inplace=True)
print(district_crime)
```

```python
bos_geo = r'Police_Districts.geojson'
bos_map = folium.Map(location=[42.3601, -71.0589], zoom_start=12)
bos_map.choropleth(
    geo_data=bos_geo,
    data=district_crime,
    columns=['DISTRICT', 'CATEGORY'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity='0.6', 
    line_opacity='0.2',
    legend_name='Crime Incident in Boston'
)
bos_map
```
